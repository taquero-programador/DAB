```python
%%sql
SELECT
    to_char(created_at, 'MM') AS month,
    COUNT(*) AS total_transactions,
    ROUND(SUM(amount_usd), 2) AS total_volume_usd
FROM company_transactions
WHERE status IN ('COMPLETE', 'PAID', 'PAYABLE')
GROUP BY month
ORDER BY month;
```
